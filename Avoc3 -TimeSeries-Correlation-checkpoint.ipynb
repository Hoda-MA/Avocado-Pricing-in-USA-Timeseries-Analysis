```python
def func_timeseries(df):
    df_kurz = df[['region','AveragePrice','Date', 'Total Volume']].copy()    
    df_kurz['Sum_total'] = df_kurz['AveragePrice'] * df_kurz['Total Volume']    
    df_grouped = df_kurz.groupby(['Date']).sum().copy()    
    df_grouped['AveragePrice'] = df_grouped['Sum_total'] / df_grouped['Total Volume']    
    daterange = pd.date_range(start=min(df['Date']), end=max(df['Date']), freq='W')
    df_ts = pd.DataFrame(daterange)
    df_ts.columns = ['Date']
    df_grouped['Date'] = df_grouped.index  
    df_grouped['Date'] =  pd.to_datetime(df_grouped['Date'])
  
    df_ts.set_index('Date') 
    df_grouped = df_grouped.reset_index(drop=True)    
    df_ts = df_ts.merge(df_grouped, on = 'Date', how='left')  
    df_ts.index = df_ts['Date']
    df_ts = df_ts.loc[:, ['AveragePrice']]   
    return df_ts
```

```python
def my_df_summary(data):
    '''Self-defined Function for summarzing'''
    try:
        dat = data.copy()
        df = pd.DataFrame([dat.sum(), dat.min(), dat.max(), dat.mean(), dat.std(), dat.isna().sum(), dat.nunique(), dat.dtypes],
                     index=['Summe', 'Minimum', 'Maximum', 'Mittelwert', 'Stand. Abw.','#NA', '#Uniques', 'dtypes'])       
        print('The No. of Total rows: {}.'.format(len(data)))
        return df
    except:
        print('No Summary Available.')
        return data
```

# Outlier Identification

```python
i_iqr_faktor = 2
df_ = df_gesamt.copy()
#df_['Date'] = pd.to_datetime(df_['Date'], format='%Y-%m-%d', errors='ignore')
df_.iloc[:,0] = np.log(df_.iloc[:,0])
df_diff = df_.diff()
q25 = df_diff.iloc[:,0].quantile(0.25)
q75 = df_diff.iloc[:,0].quantile(0.75)
iqr = q75-q25
border_down = q25 - (i_iqr_faktor*iqr)
border_over = q75 + (i_iqr_faktor*iqr)
df_[((df_diff.iloc[:,0] <  border_down) | (df_diff.iloc[:,0] >  border_over))]

df_dates = df_[((df_diff.iloc[:,:] <  border_down) | (df_diff.iloc[:,:] >border_over))]
df_dates = df_dates.reset_index()
df_dates = df_dates.iloc[:,0:]
        
print('In the entered data and the IQR factor, days on which the value change was <{0: .2f} or> {1: .2f} are noticeable.'.format (border_down, border_over))
print('This occurs on {} days: '.format(len(df_dates)))
df_dates.head()
```

# Split und Train

To test the accuracy of a model we will test the model on data that it has not seen before.
We will divide available data into two sets: a training set that the model will learn from, 
    and a test set which will be used to test the accuracy of the model on new data. 
    
   The test_size=0.2 inside the function indicates the percentage of the data that should be held over for testing. 

```python
i_split = int(0.95*len(df_gesamt))
df_train, df_test = df_gesamt.iloc[:i_split,:], df_gesamt.iloc[i_split:,:]
print('The Total entities of Test and Train: {}'.format(len(df_train)+len(df_test)))
print('The Forcasting hast a length of: {}'.format(len(df_test)))
```

```python
#Plot
#%matplotlib notebook
#plt.rcParams['figure.figsize'] = [12,6]

%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 6]

plt.plot(df_train.index, df_train.values, label='Trainingsdaten',color = 'red')
plt.plot(df_test.index, df_test.values, label='Testdaten')
plt.axvline(x = df_gesamt.index[i_split], linewidth=2, color='green', ls='--')
plt.legend(loc=2, fontsize=10)
plt.title('Price Prediction of the Total Price-List {}'.format(df_train.columns[0]), fontsize=14)
plt.xlabel('Time', fontsize=10)
plt.ylabel('Price ', fontsize=10)
plt.show()
```

# Exponential smoothing

```python
import statsmodels.api as sm
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
#Forecasts mit verschiedenen Glättungsparametern erzeugen
fit1 = SimpleExpSmoothing(df_train).fit(smoothing_level=0.2,optimized=False)
fcast1 = fit1.forecast(len(df_test)).rename(r'$\alpha=0.2$')
fit2 = SimpleExpSmoothing(df_train).fit(smoothing_level=0.75,optimized=False)
fcast2 = fit2.forecast(len(df_test)).rename(r'$\alpha=0.75$')
fit3 = SimpleExpSmoothing(df_train).fit()
fcast3 = fit3.forecast(len(df_test)).rename(r'$\alpha=%s$'%round(fit3.model.params['smoothing_level'],1))
```

```python
fit3.params
```

```python
#%matplotlib notebook
#plt.rcParams['figure.figsize'] = [12, 6]

%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 6]


plt.plot(df_train.index, df_train.values, label='Trainingsdaten',color = 'pink')
plt.plot(df_test.index, df_test.values, color='yellow', label='Testdaten')

fcast1.plot(legend=True, color='red', ls='--')
fit1.fittedvalues.plot(color='red')

fcast2.plot(legend=True, color='blue', ls='--')
fit2.fittedvalues.plot(color='blue')

fcast3.plot(legend=True, color='green', ls='--')
fit3.fittedvalues.plot(color='green')

plt.title('Adjustment of simple exponential smoothing{}'.format(df_train.columns[0]), fontsize=14)
plt.xlabel('Time', fontsize=10)
plt.ylabel('Price', fontsize=10)
plt.legend(loc=2, fontsize=12)
plt.show()
```
